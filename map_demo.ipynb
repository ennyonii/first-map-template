```python
import hvplot.pandas
import holoviews as hv
import osmnx as osm
```

```python
# Search for boundary using OSM

UR_boundary = osm.geocode_to_gdf(
                        "W155526740",
                        by_osmid=True
)

UR_boundary
```

```python
# Make an interactive plot
UR_map = UR_boundary.hvplot(
    geo=True, tiles='EsriImagery', alpha=0.5, frame_width=800, frame_height=600
)

# Save the map as a file to put on the web (OPTIONAL)
hv.save(UR_map, 'UR_map.html')

# Display the map
UR_map
```
