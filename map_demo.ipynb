```python
import hvplot.pandas
import holoviews as hv
import osmnx as osm
```

```python
# Search for boundary using OSM

cu_boundary = osm.geocode_to_gdf(
                        "W46226108",
                        by_osmid=True
)

cu_boundary
```

```python
# Make an interactive plot
cu_map = cu_boundary.hvplot(
    geo=True, tiles='EsriImagery'
)

# Save the map as a file to put on the web (OPTIONAL)
hv.save(cu_map, 'cu_map.html')

# Display the map
cu_map
```
